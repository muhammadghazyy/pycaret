## Making the model

```python
cat_var = ['sex','cp','fbs','restecg','exang','thal']
initial = pycaret.classification.setup(heart , target='target',categorical_features=cat_var)
```

```python
modelkita = pycaret.classification.compare_models()
```

```python
final = pycaret.classification.predict_model(modelkita,heart)
```

```python
submission = pd.DataFrame({"age":final['age'],"label":final['Label']})
submission
```
