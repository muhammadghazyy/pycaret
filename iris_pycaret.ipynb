```python
import pandas as pd
import pycaret.classification
import sklearn
```

```python
iris = sklearn.datasets.load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['target']=iris.target
df.head()
```

```python
iriscaret = pycaret.classification.setup(df ,target='target')
```

```python
iris_model = pycaret.classification.compare_models()
```

```python
final = pycaret.classification.predict_model(iris_model,df)
```

```python
false = final[['target','Label']].loc[final['target']!=final['Label']].count()
true = final[['target','Label']].loc[final['target']==final['Label']].count()
accuracy = (true)/(false+true)
print(f"The accuracy for the model is {accuracy}")
```
