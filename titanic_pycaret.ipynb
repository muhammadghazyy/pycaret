```python
train = pd.read_csv("train.csv")
test = pd.read_csv('test.csv')
```

```python
train1 = train.drop(['PassengerId','Age','Name','Ticket','Cabin'],axis=1)
test1 = test.drop(['PassengerId','Age','Name','Ticket','Cabin'],axis=1)

# Dropping Age , no correlation between them and survived + a lot of NaN (use df.corr())
# Dropping cabin, too much NaN
```

```python
train1.loc[train1['Embarked'].isna()]
train1
```

```python
train1 = train1.drop([61,829])
# Dropping 61 829 , dont have embarked info
```

```python
initial = pycaret.classification.setup(train1,target='Survived',categorical_features=['Pclass','Sex','SibSp','Parch','Embarked'])
```

```python
modelkita = pycaret.classification.compare_models()
```

```python
final = pycaret.classification.predict_model(modelkita,test1)
```

```python
submission = pd.DataFrame({
    'PassengerId':test['PassengerId'],
    'Survived':final['Label']
})

submission.to_csv('sub.csv',index=False)
```
